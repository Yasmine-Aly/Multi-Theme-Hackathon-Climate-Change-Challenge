```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,r2_score
```

```python
Data_1 = pd.read_csv("Data_1.csv",parse_dates=["Date"],index_col="Date")
```

```python
Data_2 = pd.read_csv("Data_2.csv")
```

```python
Data = Data_1.merge(Data_2,on=["YEAR","MO","DY"])
Data.info()
```

```python
Data.describe()
```

```python
model = LinearRegression()
```

```python
features = Data.drop(["RH2M"],axis=1)
target = Data[["RH2M"]]
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(features,target,test_size=.20,shuffle=False,random_state=0)
#p=0.8

#Xtrain = features.iloc[:int(len(features.index) * p)]
#Xtest = features.iloc[int(len(features.index) * p):]

#ytrain = target.iloc[:int(len(target.index) * p)]
#ytest = target.iloc[int(len(target.index) * p):]
```

```python
model.fit(Xtrain,ytrain)
```

```python
model.score(Xtest,ytest)
```

```python
plt.style.use('_mpl-gallery')

# plot
plt.figure(figsize=(32,16))
#
plt.plot(Data.index,Data["RH2M"],"g")
plt.plot(ytest.index,ytest["RH2M"],"r")
plt.plot(ytrain.index,ytrain["RH2M"],"b")
plt.legend(["All Data", "Test Data","Train Data"], loc ="upper left", fontsize='xx-large')
plt.show()
```

```python
MSE_TR = mean_squared_error(model.predict(Xtrain),ytrain)
MSE_TT = mean_squared_error(model.predict(Xtest),ytest)
print(MSE_TR)
print(MSE_TT)
```

```python
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(ytest, model.predict(Xtest)))
# The coefficients
print("Coefficients: \n", model.coef_)
```
